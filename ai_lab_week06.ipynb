#Prompt templates with groq llama-3.1 model

```python
import os
from google.colab import userdata

os.environ["GROQ_API_KEY"] = userdata.get("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="llama-3.1-8b-instant",  # ✔ new working model
    groq_api_key=os.environ["GROQ_API_KEY"],
    temperature=0.7
)

# questions = [
#     "what is aqi?",
#     "what is delhi aqi?"
# ]

# responses = llm.batch(questions)

# for r in responses:
#     print(r.content)
```

Converting language A to language B

```python
from langchain_core.prompts import PromptTemplate
template = PromptTemplate.from_template("Translate the following sentence from {source} to {target}: {sentence}")
```

```python
msg = template.invoke({"source": "English", "target": "Hindi", "sentence": "My name is kartikay"})
```

```python
llm.invoke(msg)
```
