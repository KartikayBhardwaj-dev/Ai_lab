```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#using airline passenger dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url, parse_dates=["Month"], index_col="Month")
df.head()
```

```python
#checking trend
df["Passengers"].plot(figsize= (12,5))
print("here is a upward trend")
print("up down season can be seen")
```

```python
#Checking reliance finance data
import yfinance as yf
ticker_symbol = "RELIANCE.NS"
stock_data = yf.download(ticker_symbol, start="2023-01-01", end="2025-08-01", interval="1d")
stock_data.columns = stock_data.columns.droplevel("Ticker")
stock_data
```

```python
#checking any trend or seasone in reliance data
stock_data["Close"].plot(figsize=(12,5))
print("we cant determine any season or trend in this")
```

```python
#use library which find trend season for us
from statsmodels.tsa.seasonal import seasonal_decompose
sd = seasonal_decompose(df["Passengers"], model="multiplicative", period=10)
trend = sd.trend
seasonal = sd.seasonal
residuals = sd.resid

```

```python
plt.figure(figsize=(14,10))

plt.subplot(411)
plt.plot(df["Passengers"])
plt.title("Original Passengers Data")

plt.subplot(412)
plt.plot(trend)
plt.title("Trend Component")

plt.subplot(413)
plt.plot(seasonal)
plt.title("Seasonal Component")

plt.subplot(414)
plt.plot(residuals)
plt.title("Residual Component")

plt.tight_layout()
plt.show()
```
